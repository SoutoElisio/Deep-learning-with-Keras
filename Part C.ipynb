```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_columns = df.columns

predictors = df[df_columns[df_columns != 'Strength']]
predictors.head()
```

```python
target = df['Strength']
target.head()
```

```python
# Normalization
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_cols = predictors_norm.shape[1]
n_cols
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python

def regression_model():
    
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
model.fit(X_train, y_train, epochs=100, verbose=2)
```

```python
loss = model.evaluate(X_test, y_test)
loss
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)

# Compute the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
```

```python
mse_list = []

# Repeat steps 1-3
for _ in range(50):
    # Step 1: Randomly split the data into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=42)
    
    # Step 2: Create and train the model on the training data
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=100, verbose=0)  # Training with 100 epochs
    
    # Step 3: Evaluate the model on the test data and compute MSE
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    
    # Append the MSE to the list
    mse_list.append(mse)

mse_list
```

```python
# mean and standard deviation of mean squared error list
mse_mean = np.mean(mse_list)
mse_std = np.std(mse_list)
```

```python
#Comparing the values, it appears that the mean MSE is significantly lower when the model is trained for a larger number of epochs;
#this suggests that the model's performance has improved with more training, which in turn indicates
#that the predictions are more consistent when the model is trained for a larger number of epochs.
#This is expected, since training a model for more epochs means it has had more opportunities to learn from the dataset,
#but the performance must always be monitored to avoid overfitting the model to the training data.
```
